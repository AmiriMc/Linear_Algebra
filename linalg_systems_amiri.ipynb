```python
import numpy as np
import matplotlib.pyplot as plt
import math
from mpl_toolkits.mplot3d import Axes3D
from IPython import display
import time
from sympy import *
```

```python
# Compute RREF of different matrices (random)
# square
# rectangular (tall, wide)
# linear dependencies (Columns, rows)

m = 5
n = 5



A = Matrix( np.random.randn(m,n) )
B = Matrix( np.random.randn(m,n) )
print('A:\n',A)
print()
print('B:\n',B)
print()

A_RREF = A.rref()
B_RREF = B.rref()
print('A_RREF:\n',A_RREF)
print()
print('B_RREF:\n',B_RREF)
print()
```
