```python
# 101. Code challenge: Implement the MCA algorithm!

#MCA algorithm in code!!!

# Compute the inverse using inv(), compare to your implementation of the MCA

# Create matrix
a = np.array([[1,3,0], [0,-2,8], [7,2,-6]])
m = a
m == a
det = np.linalg.det(a)
print(a)
print()
print('determinant:',round(det))
print()

# Create matrix of determinants
for i in range(len(a)): # rows
    # define rows and columns for submatrix
    for j in range(len(a)): # cols
        atemp = np.delete(a, i, 0) # delete row
        atemp = np.delete(atemp, j, 1) # delete col
        dets = np.linalg.det(atemp)
        m[i,j] = round(dets)
        print('i',i)
        print('j',j)
        print(atemp)
        print()

print(m)
```

```python
# Create array of matching size to original of alternating +'s and -'s 
b = np.array([[1,-1,1], [-1,1,-1], [1,-1,1]])
print(b)
print()

# Perform Hadamard multiplication (e.g. element wise) of alternating +/- matrix with matrix of determinants
c = np.multiply(m,b)
print(c)
print()
c_t = np.transpose(c)
print(c_t)
print()
ainv = c_t / 164
print(ainv)
```

```python
# original matrix x inverse matrix should result in identity matrix

#result = np.multiply(ainv,a) #ainv*a
#result = ainv*a
result = np.dot(a,ainv)
result = np.around(result)
result2 = np.linalg.inv(a)
result3 = ainv - result2

print('original matrix:\n',a)
print()
print('inverse matrix:\n', ainv)
print()
print('inverse matrix 2:\n',result2)
print()
print('subtraction of the two inverse matrices:\n',result3)
print()
print('shouuld result in identity matrix:\n', result)
```
