```python
# 68. Code challenge: rank of multiplied and summed matrices

# rules: rank of AB<=min (rank(A), rank(B))
#        rank of A+B<=rank(A) + rank(B)
# Generate two matrices (A and B), 2x5
# Compute AtA and BtB
# find their ranks
# find ranks of AtA + BtB

m = 2
n = 5

A = np.random.randn(m,n)
B = np.random.randn(m,n)
print('A:\n', A)
print()
print('B:\n', B)
print()
ra = np.linalg.matrix_rank(A)
rb = np.linalg.matrix_rank(B)
print('rank=' + str(ra))
print('rank=' + str(rb))
print()

AtA = np.matrix(A.transpose())
BtB = np.matrix(B.transpose())
print('AtA:\n', AtA)
print()
print('BtB:\n', BtB)
print()
rata = np.linalg.matrix_rank(AtA)
rbtb = np.linalg.matrix_rank(BtB)
print('rank=' + str(rata))
print('rank=' + str(rbtb))
print()

#AtAx = np.matrix(A) * AtA
#BtBx = np.matrix(B) * BtB
AtAx = np.matrix(np.dot(A, AtA))
BtBx = np.matrix(np.dot(B, BtB))
print('AtAx:\n', AtAx)
print()
print('BtBx:\n', BtBx)
AtBt = np.dot(AtAx, BtBx)
rAtBt = np.linalg.matrix_rank(AtBt)
print('rank=' + str(rAtBt))
rAtAx = np.linalg.matrix_rank(AtAx)
print('rank=' + str(rAtAx))
rBtBx = np.linalg.matrix_rank(BtBx)
print('rank=' + str(rBtBx))
print()

AtpBt = AtAx + BtBx
#AtpBt = np.matrix(AtAx) + np.matrix(BtBx)
print('AtpBt:\n', AtpBt)
rAtpBt = np.linalg.matrix_rank(AtAx+BtBx)
print('rank=' + str(rAtpBt))
print()

rank_plus = np.linalg.matrix_rank(AtAx.astype(np.float64) + BtBx.astype(np.float64))
print('rank_plus:', rank_plus)
print()

import scipy.stats as ss
rank_ss = max(ss.rankdata(AtAx+BtBx))
print(rank_ss)
print()
```

```python
# This comes up in real world problems, such as statistics problems. "Is some vector contained in the column space of 
# some matrix, which means in the span of some set of vectors?"

# Determine whether this vector
    #v = [1, 2, 3, 4]' the ' means to transform into a column vector
# is in the span of these sets
    #s = [ [4,3,6,2]T, [0,4,0,1]T ] we are thinking about this matrix as a set of column vectors
    #T = [ [1,2,2,2]T, [0,0,1,2]T ]
    
# Key here is to use the rank
    
v = [1,2,3,4]
S = np.matrix([ [4,3,6,2], [0,4,0,1]])
T = np.matrix([ [1,2,2,2], [0,0,1,2]])
S = S.transpose()
T = T.transpose()
print('S:\n', S)
print('T:\n', T)
print()

S_rank = np.linalg.matrix_rank(S)
T_rank = np.linalg.matrix_rank(T)
print('S_rank:', S_rank)
print('T_rank:', T_rank)
print()

Saug = np.column_stack((S, v))
Saug_rank = np.linalg.matrix_rank(Saug)
print(Saug)
print('Saug_rank:', Saug_rank)
print()
Taug = np.column_stack((T, v))
Taug_rank = np.linalg.matrix_rank(Taug)
print(Taug)
print('Saug_rank:', Taug_rank)
print()
```
