```python
# 146. Code challenge: reconstruct a matrix from eigenlayers

# 1. Create an m x m symmetric matrix (doesn't have to be symmetric, but makes it convenient) m = 5

# 2. Take its eigendecomposition

# 3. Show that the norm of the outer product of v_i (corresponds to one eigenvector of the matrix, can use first or last,
    # any) Compute the outer product of that eigenvector with itself and compute the norm of that eigenvector.
    
# 4. Create one layer of A as (the outer product of the eigenvector times lambda) lvv', compute its norm (norm of vv' will
    # be 1, norm of lvv' is simply going to be the eigenvalue itself) - gives you one layer of the matrix
    
# 5. Reconstruct A by summing over the eigenlayers (outer products)


# Create matrix A
m = 4
A = np.array( np.random.randn(m,m) )
A = np.dot(np.transpose(A),A) # make symmetric

# Eig of A
w, v = np.linalg.eig(A) # eigendecomposition
EigVal = w
EigVec = v
vTv_matrix = np.dot(np.transpose(EigVec), EigVec)

# show the norm of the outer product of v_i
outer = np.multiply.outer(EigVec[:,0], np.transpose(EigVec[:,0]))
rank_outer = np.linalg.matrix_rank(outer)
vEigVal = EigVec[:,0] * EigVal[0]

# Create one layer of A as lvv' , compute its norm
'''
norm = np.linalg.norm(EigVec[:,0]) # will be 1
norm = np.linalg.norm(vvT) # will be 1
'''
norm = np.dot(vEigVal, np.transpose(EigVec[:,0])) # will be corresponding eigenvalue, even w/out using norm function...
norm = np.linalg.norm(norm) # will be corresponding eigenvalue
norm_outer = np.linalg.norm(outer)

# Reconstruct A by summing over the eigenlayers (outer products)
Arecon =  np.zeros((m,m))
for i in range(0,m):
    Arecon += np.outer(EigVec[:,i], EigVec[:,i])*EigVal[i]
    rank = np.linalg.matrix_rank(Arecon)
    print("rank (loop):", rank)

A_layer = np.dot(outer, EigVal[0])
A_layer_norm = np.linalg.norm(np.dot(EigVec, np.transpose(EigVec)))

print()
print("A:\n",A)
print()
print("eigenvalues (EigVal):\n",EigVal)
print()
print("matrix of eigenvectors (EigVec):\n", EigVec)
print()
print("outer:\n", outer)
print()
print("rank of outer (should be 1):\n", rank_outer )
print()
print("norm of an eigenvector, should match corresponding eigenvalue (lambda):\n", norm)
print()
print("norm of an outer product:\n", norm_outer)
print()
print("layer of matrix A:\n", A_layer)
print()
print("layer norm of matrix A:\n", A_layer_norm)
print()
print("vTv_matrix (should be identity matrix):\n", np.round(vTv_matrix) )
print()
print("vEigVal:\n", vEigVal)
print()
print("Arecon:\n", Arecon)
print()
print("A:\n",A)
```

```python
# 148. Code challenge: trace and determinant, eigenvalues sum and product

# trace(A) = sum(eigenvalues)
# det(A) = prod(eigenvalues)
# repeat procedure for a reduced rank matrix

# full rank~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
m = 4
n = 4
A = np.array( np.random.randn(m,n) )
A = np.dot(A, np.transpose(A)) # make symmetric
w, v = np.linalg.eig(A)
tr = sum(w)
tr_1 = np.matrix.trace(A)
eig_prod = np.prod(w)
eig_prod_1 = np.linalg.det(A)

print()
print("A:\n", A)
print()
print("eigenvalues:\n", w)
print()
print("eigenvectors:\n", v)
print()
print("trace via sum of eigenvalues:\n", tr)
print()
print("trace of A via numpy function:\n", tr_1)
print()
print("determinant via product of eigenvalues:\n", eig_prod)
print()
print("determinant via numpy function:\n", eig_prod_1)
print()


# reduced rank~~~~~~~~~~~~~~~~~~~~~~~
m = 7
n = 5
A = np.array( np.random.randn(m,n) ) # Inner dimensions match, outer dimensions are bigger than inner, this will
B = np.array( np.random.randn(n,m) ) # gurantee a reduced rank matrix.
A = np.dot(A, B) # make reduced rank
#A = np.dot(A, np.transpose(A)) # make symmetric
w, v = np.linalg.eig(A)
tr = sum(w)
tr_1 = np.matrix.trace(A)
eig_prod = np.prod(w)
eig_prod_1 = np.linalg.det(A)

print()
print("A:\n", A)
print()
print("eigenvalues:\n", w)
print()
print("eigenvectors:\n", v)
print()
print("trace via sum of eigenvalues:\n", tr)
print()
print("trace of A via numpy function:\n", tr_1)
print()
print("determinant via product of eigenvalues:\n", eig_prod)
print()
print("determinant via numpy function:\n", eig_prod_1)
print()

# Outcome for reduced rank is 7 eigenvalues, but 2 are zero, thus only 5 non-zero values - rank is 5
# And of course the product of all of the eigenvalues will be zero.
```
